```python
text = "Tokenization is an important step in natural language processing. It breaks down text into meaningful units."
```

```python
# Tokenize text
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Added to download the missing resource
from nltk.tokenize import word_tokenize

word_tokens = word_tokenize(text)
print("Word-level tokens:", word_tokens)
```

```python
# Tokenize text (with spacy)
import spacy

nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

word_tokens = [token.text for token in doc]
print("Word-level tokens:", word_tokens)
```

```python
# Character tokenization
char_tokens = list(text)
print("Character-level tokens:", char_tokens)
```

```python
# BPE tokenization

from tokenizers import Tokenizer, models, trainers, pre_tokenizers

# Initialize BPE
tokenizer = Tokenizer(models.BPE())
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()
trainer = trainers.BpeTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])

# Train on your text
tokenizer.train_from_iterator([text], trainer)

output = tokenizer.encode(text)
print("BPE tokens:", output.tokens)
```

```python
# word piece
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer.tokenize(text)
print("WordPiece tokens:", tokens)
```

```python
#sentence piece

import sentencepiece as spm

# Save text to a temporary file
with open("sample.txt", "w") as f:
    f.write(text)

# Train SentencePiece model
spm.SentencePieceTrainer.Train('--input=sample.txt --model_prefix=spm_model --vocab_size=36') # Changed vocab_size to 36

# Load and encode
sp = spm.SentencePieceProcessor()
sp.Load("spm_model.model")
tokens = sp.EncodeAsPieces(text)
print("SentencePiece tokens:", tokens)
```

# Tokenization Techniques

 ## 1. Word-Level Tokenization:

Splits text into words and punctuation.

**Observation:**

Simple and intuitive

Large vocabulary

Cannot handle unseen words effectively

#2. Character-Level Tokenization

Splits text into individual characters.

**Observation:**

Very small vocabulary

Long token sequences

Weak semantic representation

# 3. Subword-Level Tokenization

**i**. **byte Pair Encoding (BPE)**

Merges frequent character pairs into subwords

Handles rare words better than word-level tokenization


**ii**. **WordPiece**

Used in models like BERT

Uses ## to indicate subword continuation

Efficient handling of unknown words


iii. **SentencePiece**

Language-agnostic

Does not rely on whitespace

Suitable for multilingual tasks

# Word-Level Tokenization
**Advantages:**

Simple and intuitive

Easy to interpret and implement

**Limitations:**

Results in a very large vocabulary

Cannot effectively handle unseen or rare words

# Character-Level Tokenization


**Advantages**:

Can represent any word, including unseen or misspelled ones

Very small and fixed vocabulary

**Limitations:**

Produces very long token sequences

Makes it harder for models to capture semantic meaning

#Byte Pair Encoding (BPE)

Byte Pair Encoding is a subword tokenization technique that iteratively merges the most frequent character pairs to form subword units.

**Advantages:**

Reduces vocabulary size

Handles rare and compound words effectively

**Limitations:**

Requires training on a dataset

Token splits depend on training data distribution

# WordPiece Tokenization

Another subword tokenization method. it selects subwords based on maximizing likelihood and is commonly used in transformer models such as BERT.

**Advantages:**

Efficient handling of unknown words

Widely used in state-of-the-art NLP models

**Limitations:**

Subword tokens can be less interpretable

Requires a pretrained vocabulary

# SentencePiece Tokenization

Language-independent subword tokenization technique that does not rely on whitespace for word segmentation.

**Advantages:**

Suitable for multilingual and low-resource languages

Does not require pre-tokenized text

**Limitations:**

Output tokens may appear complex

Less intuitive for human interpretation
